```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
def grade(score):
  if score >= 90 and score <= 100:
        return 'AA'
  elif score >= 80 and score <= 89.9:
        return 'BA'
  elif score >= 70 and score <= 79.9:
        return 'BB'
  elif score >= 65 and score <= 69.9:
        return 'CB'
  elif score >= 60 and score <= 64.9:
        return 'CC'
  elif score >= 50 and score <= 59.9:
        return 'DC'
  elif score >= 30 and score <= 49.9:
        return 'DD'
  else:
        return 'FF'
```

```python
def success(score):
  if score >= 60 and score <= 100:
      return 'Succesful'
  else:
      return 'Fail'
```

```python
df = pd.read_excel("/content/drive/MyDrive/GLOBAL - Bootcamp/Sınav_Notları.xlsx")
df["NO"] = df["NO"].astype(int)
df
```

##**Results for existing student**

```python
def existing():
  global df
  no = input("Please enter the number of the student: ")
  midterm = input("Midterm: ")
  final = input("Final: ")

  score = (int(midterm)*30 + int(final)*70)/100
  print("Score: ",score)

  student = df.query(f'NO == {no}')
  student["Midterm"] = midterm
  student["Final"] = final
  student["Score"] = score
  student["Grade"] = grade(score)
  student["Success"] = success(score)

  df = df.append(student, ignore_index=True)
  return df
```

```python
#drop NaN values and then reset the dataframe's index
def drop_na_values():
  global df
  df.dropna(subset = ['Success'], inplace=True)
  df = df.reset_index(drop=True)
  return df
```

##**Adding a new student**

```python
def new_student():
  global df
  no = input("Please enter the number of the student: ")
  full_name = input("Student's fullname: ")
  midterm = input("Midterm: ")
  final = input("Final: ")

  score = (int(midterm)*30 + int(final)*70)/100
  print("Score: ",score)

  new = pd.Series(data=[no, full_name, midterm, final, score, grade(score), success(score)], index=['NO', 'Fullname','Midterm', 'Final', 'Score', 'Grade', 'Success'])
  df = df.append(new, ignore_index=True)
  return df
```
